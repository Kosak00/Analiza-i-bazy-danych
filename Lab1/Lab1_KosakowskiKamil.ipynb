```python
import numpy as np
import scipy
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# Zad.3
#Implementacja funkcji 
def function(x):
  return x*x + 5

#Wykres dla x>-1 oraz x<1 
x1 = np.linspace(-1, 1)
y1 = function(x1)
plt.plot(x1, y1)
plt.title("Wykres funkcji w przedziale (-1, 1)")
plt.legend(["f(x) = x^2 + 5"])
plt.xlabel("x")
plt.ylabel('y')
plt.grid()
plt.show()

#Wykres dla x>-6 oraz x<6 
x2 = np.linspace(-6,6)
y2 = function(x2)
plt.plot(x2, y2)
plt.title("Wykres funkcji w przedziale (-6, 6)")
plt.legend(["f(x) = x^2 + 5"])
plt.xlabel("x")
plt.ylabel('y')
plt.grid()
plt.show()

#Wykres dla x>0 oraz x<5 
x3 = np.linspace(0,5)
y3 = function(x3)
plt.plot(x3, y3)
plt.title("Wykres funkcji w przedziale (0, 5)")
plt.legend(["f(x) = x^2 + 5"])
plt.xlabel("x")
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
# Zad.4
data = np.array([("Kamil", "Kosakowski", 22, "man"),
                 ("Max", "Verstappen", 25, "man"),
                 ("Iga", "Świątek", 21, "woman"),
                 ("Lando", "Norris", 22, "man"),
                 ("Zuzanna", "Górecka", 22, "woman")
                 ])

# Utworzenie data frame z rekordów
data_frame = pd.DataFrame.from_records(data, columns=['Name', 'Surname', 'Age', 'Sex'])

# Użycie funkcji info
info = data_frame.info(verbose=True)
print(info, f'\n')

info2 = data_frame.info(verbose=False)
print(info2, f'\n')

# Użycie funkcji describe 
describe = data_frame.describe()
print(describe, f'\n')

# Wypisanie 3 pierwszych rekordów
head3 = data_frame.head(3)
print(head3)
```
